**Correlation among the columns**

```python
non_numeric_cols = df_orders.select_dtypes(exclude=["int64", "float64"]).columns.tolist()
```

```python
df_orders_numeric = df_orders.drop(non_numeric_cols, axis=1)
```

```python
df_orders["Ship Mode"] = df_orders["Ship Mode"].astype("category").cat.codes
```

```python
df_orders_numeric.corr()
```

```python
# corr() function is to find the correlation between columns.we are having four numeric column in dataframe. corr() is used to find the pairwise correlation of all columns in the dataframe.Null values are automaticaly excluded. for any non-numeric datatype columns in the dataframe is ignored
```

**Covarience between set of variables**

```python
df_orders_numeric.cov()
```

```python
# Covarience provides the measure of strength of correlation between two variables or more set os variables
# The covarience matrix element cij is the covarience of Xi and Xj.The element of Cii is the varience of Xi
#if COV(Xi,Xj)=0 then variables are not correlated
#if COV(Xi,Xj)>0 then variables are positively correlated
#if COV(Xi,Xj)> <0 then variables are negatively correlated.
```
